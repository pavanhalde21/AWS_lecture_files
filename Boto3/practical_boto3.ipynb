```python
import boto3
import yaml
import os

# Handle __file__ not defined in Jupyter
try:
    base_dir = os.path.dirname(__file__)
except NameError:
    base_dir = os.getcwd()  # current working directory

# Path to YAML file
CONFIG_PATH = os.path.join(base_dir, "config", "config.yml")

print(CONFIG_PATH)
# Load YAML config
with open(CONFIG_PATH, "r") as f:
    config = yaml.safe_load(f)

aws_conf = config["aws"]

# Create S3 client using credentials from YAML
s3_client = boto3.client(
    "s3",
    aws_access_key_id=aws_conf["access_key_id"],
    aws_secret_access_key=aws_conf["secret_access_key"],
    region_name=aws_conf["region"]
)

# Example: list buckets
response = s3_client.list_buckets()
# print(response)
for bucket in response["Buckets"]:
    print(bucket["Name"])
```

# AWS GLUE

```python
# Create an glue client
glue_client = boto3.client('glue', 
                  aws_access_key_id='',
                  aws_secret_access_key='',
                  region_name='us-east-1')
```

```python
# Listing jobs in account:
response = glue_client.get_jobs()
# print(response)
jobs=response['Jobs']
for job in jobs:
    print(f"Job Name: {job['Name']}")
```

```python
# Start Job Run ::
response = glue_client.start_job_run(JobName='glue_job_02')
job_run_id = response['JobRunId']
print(f"Job run started with ID: {job_run_id}")
```

```python
# Get Job Run::
response = glue_client.get_job_run(JobName='glue_job_02', RunId=job_run_id)
job_run = response['JobRun']
print(f"Job Run Status: {job_run['JobRunState']}")
```

```python
# List Crawlers ::
response = glue_client.get_crawlers()
crawlers = response['Crawlers']
for crawler in crawlers:
    print(f"Crawler Name: {crawler['Name']}, Crawler State: {crawler['State']}")
```

```python
# Start Crawler::
response = glue_client.start_crawler(Name='crawler_01')
print(response)
```

```python
name = 'crawler_01'
resp = glue_client.get_crawler(Name=name)
c = resp.get("Crawler", {})
state = c.get("State")
last_crawl = c.get("LastCrawl", {})
last_status = last_crawl.get("Status")
start_time = last_crawl.get("StartTime")
print(f"Crawler: {name}")
print(f"State: {state}")
if last_crawl:
    print(f"Last crawl status: {last_status}")
    print(f"Last crawl start time: {start_time}")
else:
    print("No last crawl information available.")



```

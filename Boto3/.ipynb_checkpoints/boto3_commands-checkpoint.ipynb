# S3

```python
# Create an S3 client
s3_client = boto3.client('s3', 
                  aws_access_key_id=fetch_credentials()[0],
                  aws_secret_access_key=fetch_credentials()[1],
                  region_name='us-east-1')
```

```python
# Listing buckets in AWS Account with above access and secret access key 
s3_client.list_buckets()
```

```python
# Bucket Creation:
s3_client.create_bucket(Bucket='cab-boto3-demo1')
```

```python
# upload file to S3:
# s3_client.upload_file(file_path, bucket_name, object_name,ExtraArgs)
s3_client.upload_file(r"C:\Users\ADMIN\Desktop\AWSDSBatch\Syallabus_Coverage.txt", 'cab-boto3-demo', 'aws_syllabus')

#Explore ExtraArgs parameter(optional) 

# Explore "upload_fileobj" method
```

```python
# Downloading file from S3:
# s3_client.download_file(bucket_name, object_key, download_path)
s3_client.download_file('cab-data-science-demo', 'EMR/input/customer.csv', r'C:\Users\ADMIN\Desktop\AWSDSBatch\boto3\cust.csv')
```

```python
##Bucket Policy:
# 1] get_bucket_policy:
result = s3_client.get_bucket_policy(Bucket='cab-glue')
print(result['Policy'])
```

```python
# 2]Set a bucket policy:
import json
bucket_policy={
    "Version": "2012-10-17",
    "Statement": [
        {
            "Effect": "Allow",
            "Principal": "*",
            "Action": "s3:GetObject",
            "Resource": "arn:aws:s3:::cab-data-science-demo/*"
        }
    ]
}
# Convert the policy from JSON dict to string
bucket_policy_string = json.dumps(bucket_policy)
s3_client.put_bucket_policy(Bucket='cab-data-science-demo', Policy=bucket_policy_string)
```

```python
# Delete Bucket Policy:
s3_client.delete_bucket_policy(Bucket='cab-data-science-demo')
```

```python
# Get a bucket access control list:
result = s3_client.get_bucket_acl(Bucket='cab-glue')
print(result)
```

# AWS GLUE

```python
# Create an glue client
glue_client = boto3.client('glue', 
                  aws_access_key_id=fetch_credentials()[0],
                  aws_secret_access_key=fetch_credentials()[1],
                  region_name='us-east-1')
```

```python
# Listing jobs in account:
response = glue_client.get_jobs()
jobs=response['Jobs']
for job in jobs:
    print(f"Job Name: {job['Name']}")
```

```python
# Start Job Run ::
response = glue_client.start_job_run(JobName='t75job')
job_run_id = response['JobRunId']
print(f"Job run started with ID: {job_run_id}")
```

```python
# Get Job Run::
response = glue_client.get_job_run(JobName='t75job', RunId=job_run_id)
job_run = response['JobRun']
print(f"Job Run Status: {job_run['JobRunState']}")
```

```python
# List Crawlers ::
response = glue_client.get_crawlers()
crawlers = response['Crawlers']
for crawler in crawlers:
    print(f"Crawler Name: {crawler['Name']}, Crawler State: {crawler['State']}")
```

```python
# Start Crawler::
response = glue_client.start_crawler(Name='Garage_crawler')
print(response)
```

```python
# Delete Job ::
glue_client.delete_job(JobName='presc_job')
print("Job deleted")
```

# AWS EMR

```python
emr_client=boto3.client('emr', 
                  aws_access_key_id=fetch_credentials()[0],
                  aws_secret_access_key=fetch_credentials()[1],
                  region_name='us-east-1')
```

```python
# Create EMR cluster
response = emr_client.run_job_flow(
    Name='ds-cluster-boto3',
    ReleaseLabel='emr-6.14.0',
    Instances={
        'MasterInstanceType': 'm5.xlarge',
        'SlaveInstanceType': 'm5.xlarge',
        'InstanceCount': 2,
        'KeepJobFlowAliveWhenNoSteps': True,
        'TerminationProtected': False,
        'Ec2KeyName': 'first_keypair',
        'EmrManagedMasterSecurityGroup': 'sg-0309ac88be23385a2',
        'EmrManagedSlaveSecurityGroup': 'sg-0a19af4ba0a963cca',
        'Ec2SubnetId': 'subnet-090cd51dd31d81d26',
    },
    Applications=[
        {'Name': 'Spark'},
    ],
    VisibleToAllUsers=True,
    JobFlowRole='arn:aws:iam::182424271996:instance-profile/EMR_EC2_DefaultRole', #this is instance profile role
    ServiceRole='arn:aws:iam::182424271996:role/EMR_Default_Role',
)
print(response)
```

```python
# Add steps to the cluster 
step_name = 'MySparkJob'
jar_location = 's3://your-bucket/your-spark-job.jar'
main_class = 'com.example.YourSparkJobClass' 
arguments = ['arg1', 'arg2']  

step_response = emr_client.add_job_flow_steps(
    JobFlowId=cluster_id,
    Steps=[
        {
            'Name': step_name,
            'ActionOnFailure': 'CONTINUE',
            'HadoopJarStep': {
                'Jar': 'command-runner.jar',
                'Args': ['spark-submit', main_class, jar_location] + arguments,
            },
        },
    ]
)

# spark-submit command: spark submit,--deploy-mode,'cluster',--master
step_id = step_response['StepIds'][0]
print(f"Step added with ID: {step_id}")
```

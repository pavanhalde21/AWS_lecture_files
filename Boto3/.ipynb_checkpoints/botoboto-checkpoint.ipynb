```python
import boto3

# Initialize an S3 client
s3 = boto3.client('s3')

# List objects in a bucket
response = s3.list_objects(Bucket='pavan-halde-glue-1')
```

```python
response
```

```python
# Print the object keys
for obj in response['Contents']:
    print('Object Key: ' + obj['Key'])
```

```python
# creating a bucket
s3_client = boto3.client('s3')
s3_client.create_bucket(Bucket="pavan-halde-bucket-boto3")
```

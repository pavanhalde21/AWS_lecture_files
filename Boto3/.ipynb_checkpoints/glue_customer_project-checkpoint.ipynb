```python
import boto3

glue_client = boto3.client('glue')
```

```python
# Create a database in the Data Catalog
response = glue_client.create_database(
    DatabaseInput={
        'Name': 'pavan-glue-customer-boto3-database'
    }
)

```

```python
# Create the Crawler
crawler_name = 'pavan-customer-boto3-crawler'

s3_path1 = 's3://pavan-emr-boto3-customer/input/customer/customer/'
s3_path2 = 's3://pavan-emr-boto3-customer/input/sales/sales/'

response = glue_client.create_crawler(
    Name=crawler_name,
    Role='arn:aws:iam::032731649829:role/pavancrawlerfetchfromS3',
    DatabaseName='pavan-glue-customer-boto3-database',  
    Targets={
        'S3Targets': [
            {
                'Path': s3_path1
            },
            {
                'Path': s3_path2  
            }
        ]
    }
)
```

```python
# Start the Crawler
response = glue_client.start_crawler(
    Name=crawler_name
)
```

```python
# List the tables created by the Crawler
response = glue_client.get_tables(DatabaseName='pavan-glue-customer-boto3-database')
tables = response['TableList']
for table in tables:
    print(f"Table Name: {table['Name']}")
```

```python
response = glue_client.create_job(
    Name='pavan-boto3-glue-customer-job1',
    Role='arn:aws:iam::032731649829:role/pavancrawlerfetchfromS3',
    Command={
        'Name': 'glueetl',
        'ScriptLocation': 's3://pavan-emr-boto3-customer/glue-script/glue_customer.py'
    },
    GlueVersion='4.0',
    WorkerType='G.1X',
    NumberOfWorkers=2,
    ExecutionProperty={
        'MaxConcurrentRuns': 1
    }
)
```

```python
# running job
response = glue_client.start_job_run(
    JobName='pavan-boto3-glue-customer-job1'
)
```

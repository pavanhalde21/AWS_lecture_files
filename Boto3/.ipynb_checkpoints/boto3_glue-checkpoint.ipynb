```python
import boto3

glue_client = boto3.client('glue')
```

```python
# Create a database in the Data Catalog
response = glue_client.create_database(
    DatabaseInput={
        'Name': 'pavan-boto3-database'
    }
)

# Create a table in the database
response = glue_client.create_table(
    DatabaseName='pavan-boto3-database',
    TableInput={
        'Name': 'product',
        # Define the table structure and other properties here
    }
)
```

```python
# Define the name of your Crawler
crawler_name = 'pavan-boto3-crawler'

# Specify the S3 path to your data
s3_path = 's3://pavan-halde-glue-etl-project1-boto3-copy/input/product/'

# Create the Crawler
response = glue_client.create_crawler(
    Name=crawler_name,
    Role='arn:aws:iam::032731649829:role/pavancrawlerfetchfromS3',
    DatabaseName='pavan-boto3-database',  # The database where you want to catalog the data
    Targets={
        'S3Targets': [
            {
                'Path': s3_path
            }
        ]
    }
)

```

```python
# Start the Crawler
response = glue_client.start_crawler(
    Name=crawler_name
)
```

```python
# Wait for the Crawler to finish (optional)
crawler_state = None
while crawler_state not in ('READY', 'STOPPING'):
    response = glue_client.get_crawler(Name=crawler_name)
    crawler_state = response['Crawler']['State']
    print(f"Crawler State: {crawler_state}")
```

```python
# List the tables created by the Crawler
response = glue_client.get_tables(DatabaseName='pavan-boto3-database')
tables = response['TableList']
for table in tables:
    print(f"Table Name: {table['Name']}")
    
```

```python

response = glue_client.create_job(
    Name='pavan-boto3-glue-job',
    Role='arn:aws:iam::032731649829:role/pavancrawlerfetchfromS3',
    Command={
        'Name': 'glueetl',
        'ScriptLocation': 's3://pavan-halde-glue-etl-project1-boto3-copy/scripts/pavanGlueJobReadFromS3.py'
    },
    GlueVersion='4.0',
    WorkerType='G.1X',
    NumberOfWorkers=2,
    ExecutionProperty={
        'MaxConcurrentRuns': 1
    }
)
```
